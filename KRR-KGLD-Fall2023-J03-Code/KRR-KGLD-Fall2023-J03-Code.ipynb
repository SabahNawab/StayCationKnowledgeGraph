```python
grph= Graph()
grph.parse(r"D:\krr\KRR-Fall2023-Project03-J03\KRR-Fall2023-Project03-J03_FILE2.rdf", format="xml")
stay=Namespace("http://staycation.org#")
schema=Namespace("http://schema.org/")
acco=Namespace("http://purl.org/acco/ns#")
```

# DATAFILES 

```python
for index, row in room_type.iterrows():
    if row['Room_Name'] == "Suite":
        room_name = URIRef(acco + row['Room_Name'].replace(" ", "_"))
    else:
        room_name = URIRef(stay + row['Room_Name'].replace(" ", "_"))

    room_type_value = URIRef(stay + row['Room_Type'].replace(" ", "_"))
    view=URIRef(stay + row['ViewType'].replace(" ", "_"))
    grph.add((room_name, stay.hasViewType, view))
#     grph.add((room_name, RDFS.label, Literal(row['Room_Name'],datatype=XSD.string)))
    grph.add((view, RDFS.label, Literal(row['ViewType'],datatype=XSD.string)))
    grph.add((view, RDF.type, stay.ViewType))
    grph.add((room_name, stay.hasCapacityAdults, Literal(row['Adult'], datatype=XSD.integer)))
    grph.add((room_name, stay.hasCapacityChildren, Literal(row['Child'], datatype=XSD.integer)))
    grph.add((room_name, RDFS.subClassOf, room_type_value))
    grph.add((room_name, RDFS.label, Literal(row['Room_Name'],datatype=XSD.string)))
    grph.add((room_type_value, RDFS.subClassOf, schema.Room)) 
    grph.add((room_name, stay.hasRoomID, Literal(row['Room_ID'], datatype=XSD.integer)))
```

```python
hotels = pd.read_csv(r"D:\Hotel - hotels.csv")
for index, row in hotels.iterrows():
    
    hotel_name = URIRef(stay + row['Name'].replace(" ", "_"))
    city_name = URIRef(stay + row['City'].replace(" ", "_"))
    country_name = URIRef(stay + row['Country'].replace(" ", "_"))
    brand_name = URIRef(stay + row['Brand Name'].replace(" ", "_"))
    
    
    grph.add((hotel_name, stay.isPartOfCity, city_name))
    grph.add((hotel_name, stay.hasClosestLandMark, Literal(row['Nearest Landmark'], datatype=XSD.string)))
    grph.add((hotel_name,RDF.type,acco.Hotel))
    grph.add((hotel_name,RDFS.label,Literal(row['Name'],datatype=XSD.string)))
    grph.add((city_name, stay.isPartOfCountry, country_name))
    grph.add((country_name, stay.hasCountryCode, Literal(row['Country Code'], datatype=XSD.string)))
    grph.add((city_name, stay.hasCityID, Literal(row['City ID'], datatype=XSD.integer)))
    grph.add((hotel_name, stay.hasRatingStar, Literal(row['Rating Stars'], datatype=XSD.float)))
    grph.add((hotel_name, stay.hasRatingScore, Literal(row['Rating Score'], datatype=XSD.float)))
    grph.add((hotel_name, stay.hasCheckOutTime, Literal(row['Check out'], datatype=XSD.datetime)))
    grph.add((hotel_name, stay.hasCheckInTime, Literal(row['Check In'], datatype=XSD.datetime)))
    grph.add((hotel_name, stay.hasBrand, brand_name))
    grph.add((hotel_name, stay.hasAccomodationID, Literal(row['Hotel ID'], datatype=XSD.integer)))
    grph.add((brand_name, stay.hasBrandID, Literal(row['Brand ID'], datatype=XSD.integer)))
```
